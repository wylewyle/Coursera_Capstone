```python
import pandas as pd
import numpy as np
```

```python
#Import the Beautiful Soup package
import requests
from bs4 import BeautifulSoup as BS
```

```python
rawdat=requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
```

```python
rawdat
```

```python
soupdat=BS(rawdat,'xml')
```

```python
soupdat
```

```python
souptable=soupdat.find('table')
```

```python
souptable
```

```python
#It's tr and td that we need to look out for
#Set up dataframe

PBN = pd.DataFrame(columns = ['PostalCode','Borough','Neighborhood'])

for TR in souptable.find_all('tr'):
    #initiatlize empty rowlist for every iteration of rows
    listfill=[]
    for TD in TR.find_all('td'):
        listfill.append(TD.text.strip())
    #stop when full
    if len(listfill)>2:
        PBN.loc[len(PBN)] = listfill
PBN.head()
```

```python
PBNcleaned=PBN.drop(PBN[PBN.Borough == 'Not assigned'].index, inplace=False)
```
